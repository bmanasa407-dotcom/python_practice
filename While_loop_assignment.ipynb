Problem 10.1: Print 1 to N Using While (Easy)
Print numbers from 1 to n using while loop.

Hint: Initialize i=1. While i <= n: print(i), i += 1.

```python
n = int(input("enter a number :"))
i = 1
while i <= n:
    print(i)
    i +=1
```

Problem 10.2: Countdown Timer (Easy)
Create a countdown from n to 1 using while loop.

Hint: While n > 0: print(n), n -= 1

```python
n = int(input("Enter a number: "))


while n > 0:
    print(n)
    n -= 1  
```

Problem 10.3: Find First Digit (Easy)
Find the first digit of a number using while loop.

Hint: While n >= 10: n = n // 10. When loop ends, n is the first digit.

```python
n = int(input("Enter a number: "))
while n > 10:
    n= n//10
print(n)
```

Problem 10.7: Fibonacci Until N (Medium)
Print Fibonacci numbers until they exceed n.

Hint: a=0, b=1. While a <= n: print(a), calculate next (a, b = b, a+b).

```python
n = int(input("Enter a number: "))

a, b = 0, 1


while a <= n:
    print(a)
    a, b = b, a + b   
```

Problem 10.9: Factorial Using While (Medium)
Calculate factorial using while loop.

Hint: result=1, i=1. While i <= n: result *= i, i += 1.

```python
n = int(input("Enter a number: "))

result = 1
i = 1


while i <= n:
    result *= i   
    i += 1        

print("Factorial of", n, "is:", result)
```

```python
x = int(input("enter the base number :"))
n = int(input("enter the exponent number :"))
count = 0
result = 1
while count < n:
    result *= x
    count +=1
print (result)
```
